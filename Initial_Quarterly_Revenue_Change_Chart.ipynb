```python
import matplotlib.pyplot as plt 
plt.rcdefaults()
import pandas as pd
import numpy as np

#import the data set from VerizonFinancialsCleaned.csv file
data = pd.read_csv('VerizonFinancialsCleaned.csv', sep=',',header=None, index_col =0)

quarterly_rev_change = []   #this list holds the quarterly_rev_change values
quarters = []  
i = 0
j=0

#creater a list for quarters
for m in range(1,49):
    quarters.append(m)

#calculater the quarterly reveneu changes
for r in range(1,49):
    i=i+1
    j = i+1
    if(j<48):
        quarterly_rev_change.append(int(data[7][j]) - int(data[7][i]) )
    else:
        break ; 

print(quarterly_rev_change)
        
#draw a chart 
y_pos = np.arange(len(quarterly_rev_change))

plt.bar(y_pos, quarterly_rev_change, alpha=1)
plt.xticks(y_pos, quarters)
plt.ylabel('Change in Quarterly Revenue')
plt.xlabel('Quarters')
 
plt.show()
```
