```python
# check for all rows with nan values
nan = data2[data2.isnull().any(axis=1)]
nan
```

```python
#Create new dataframe, copy dates, normalize all columns
#Perform lognormal transformation for rest of the data
# lognormal = percentage change per day, but can add or subtract cumulatively, unlike percentage
# multiple values x 100 in your head to get the "percentage"

data2norm = data2.filter(['Date'], axis=1)
data2norm['CMCSA']=np.log(data2.CMCSA/data2.CMCSA.shift(1))
data2norm['VZ']=np.log(data2.VZ/data2.VZ.shift(1))
data2norm['TMUS']=np.log(data2.TMUS/data2.TMUS.shift(1))
data2norm['ATT']=np.log(data2.ATT/data2.ATT.shift(1))
data2norm['QCOM']=np.log(data2.QCOM/data2.QCOM.shift(1))
data2norm['QQQ']=np.log(data2.QQQ/data2.QQQ.shift(1))
data2norm['SPY']=np.log(data2.SPY/data2.SPY.shift(1))
data2norm['IXP']=np.log(data2.IXP/data2.IXP.shift(1))
data2norm['IYZ']=np.log(data2.IYZ/data2.IYZ.shift(1))
data2norm['IXTC']=np.log(data2.IXTC/data2.IXTC.shift(1))

#check data
data2norm.head()
```

```python
#Check dataframe for nan values
nan = data2norm[data2norm.isnull().any(axis=1)]
nan
```

```python
# need to drop first row
# reset index numbers
data2norm=data2norm.drop(data2norm.index[0])
data2norm.index  = range(len(data2norm.index))
data2norm.head()
```

```python
# create data sets for comparison to indices
# compares normalized data only (not price)
data2Comp = data2norm.filter(['Date'], axis=1)
# QQQ
data2Comp['CMSCA-QQQ'] = data2norm['CMCSA']-data2norm['QQQ']
data2Comp['VZ-QQQ'] = data2norm['VZ']-data2norm['QQQ']
data2Comp['TMUS-QQQ'] = data2norm['TMUS']-data2norm['QQQ']
data2Comp['ATT-QQQ'] = data2norm['ATT']-data2norm['QQQ']
data2Comp['QCOM-QQQ'] = data2norm['CMCSA']-data2norm['QQQ']
# SPY
data2Comp['CMSCA-SPY'] = data2norm['CMCSA']-data2norm['SPY']
data2Comp['VZ-SPY'] = data2norm['VZ']-data2norm['SPY']
data2Comp['TMUS-SPY'] = data2norm['TMUS']-data2norm['SPY']
data2Comp['ATT-SPY'] = data2norm['ATT']-data2norm['SPY']
data2Comp['QCOM-SPY'] = data2norm['CMCSA']-data2norm['SPY']
# IXP
data2Comp['CMSCA-IXP'] = data2norm['CMCSA']-data2norm['IXP']
data2Comp['VZ-IXP'] = data2norm['VZ']-data2norm['IXP']
data2Comp['TMUS-IXP'] = data2norm['TMUS']-data2norm['IXP']
data2Comp['ATT-IXP'] = data2norm['ATT']-data2norm['IXP']
data2Comp['QCOM-IXP'] = data2norm['CMCSA']-data2norm['IXP']
# IYZ
data2Comp['CMSCA-IYZ'] = data2norm['CMCSA']-data2norm['IYZ']
data2Comp['VZ-IYZ'] = data2norm['VZ']-data2norm['IYZ']
data2Comp['TMUS-IYZ'] = data2norm['TMUS']-data2norm['IYZ']
data2Comp['ATT-IYZ'] = data2norm['ATT']-data2norm['IYZ']
data2Comp['QCOM-IYZ'] = data2norm['CMCSA']-data2norm['IYZ']
# IXTC
data2Comp['CMSCA-IXTC'] = data2norm['CMCSA']-data2norm['IXTC']
data2Comp['VZ-IXTC'] = data2norm['VZ']-data2norm['IXTC']
data2Comp['TMUS-IXTC'] = data2norm['TMUS']-data2norm['IXTC']
data2Comp['ATT-IXTC'] = data2norm['ATT']-data2norm['IXTC']
data2Comp['QCOM-IXTC'] = data2norm['CMCSA']-data2norm['IXTC']
data2Comp.head()
```

```python
# create csv file of final data for future projects
data2norm.to_csv('TelcomNorm.csv')
data2Comp.to_csv('TelecomComparison.csv')
```
