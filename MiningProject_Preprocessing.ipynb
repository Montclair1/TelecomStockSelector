```python
# Load Pandas libraries with alias 'pd' 
import pandas as pd 
data1 = pd.read_csv("VerizonFinancialsCleaned.csv")
data2 = pd.read_csv("Telcom.csv") 
data1.head()
```

```python
data2.head()
```

```python
#TEST ONE COLUMN
#Create new columns for daily lognormal return
import numpy as np
# normalize data with natural log of difference between each day
data2['Tlog']=np.log(data2.CMCSA/data2.CMCSA.shift(1))
data2.head()
```

```python
# check for all rows with nan values
nan = data2[data2.isnull().any(axis=1)]
nan
```

```python
#Create new dataframe, copy dates, normalize all columns
#Perform lognormal transformation for rest of the data
# lognormal = percentage change per day, but can add or subtract cumulatively, unlike percentage
# multiple values x 100 in your head to get the "percentage"

data2norm = data2.filter(['Date'], axis=1)
data2norm['CMCSA']=np.log(data2.CMCSA/data2.CMCSA.shift(1))
data2norm['VZ']=np.log(data2.VZ/data2.VZ.shift(1))
data2norm['TMUS']=np.log(data2.TMUS/data2.TMUS.shift(1))
data2norm['T']=np.log(data2.CMCSA/data2.CMCSA.shift(1))
data2norm['QCOM']=np.log(data2.QCOM/data2.QCOM.shift(1))
data2norm['QQQ']=np.log(data2.QQQ/data2.QQQ.shift(1))
data2norm['SPY']=np.log(data2.SPY/data2.SPY.shift(1))
data2norm['IXP']=np.log(data2.IXP/data2.IXP.shift(1))
data2norm['IYZ']=np.log(data2.IYZ/data2.IYZ.shift(1))
data2norm['IXTC']=np.log(data2.IXTC/data2.IXTC.shift(1))

#check data
data2norm.head()
```

```python
#Check dataframe for nan values
nan = data2norm[data2norm.isnull().any(axis=1)]
nan
```

```python
# need to drop first row
data2norm.drop(data2norm.index[0])
```

```python
# create csv file of final data for future projects
data2norm.to_csv('TelcomNorm.csv')
```
