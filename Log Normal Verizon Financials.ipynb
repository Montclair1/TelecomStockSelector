```python
import pandas as pd
data1 = pd.read_csv("VerizonFinancialsCleaned.csv")
data1.head()
```

```python
import numpy as np
# normalize data with natural log of difference between each day, add as extra row
data1['ln StockPrice']=np.log(data1.StockPrice/data1.StockPrice.shift(1))
data1.head()
```

```python
#create new dataframe copy the columns and normalize the StockPrice column
data1norm = data1.filter(['Date'], axis=1)
data1norm['StockPrice'] = data1['StockPrice']
data1norm['Log Normal StockPrice']=np.log(data1.StockPrice/data1.StockPrice.shift(1))
data1norm['TTMNetEPS'] = data1['TTMNetEPS']
data1norm['TTMSalesperShare'] = data1['TTMSalesperShare']
data1norm['PricetoSalesRatio'] = data1['PricetoSalesRatio']
data1norm['BookValueperShare'] = data1['BookValueperShare']
data1norm['TTMFCFperShare'] = data1['TTMFCFperShare']
data1norm['QuarterlyRevenue'] = data1['QuarterlyRevenue']
data1norm['GrossProfit'] = data1['GrossProfit']
data1norm['CurrentAssetsInventory'] = data1['CurrentAssetsInventory']
data1norm['OperatingIncome'] = data1['CurrentAssetsInventory']
data1norm['QuickRatio'] = data1['QuickRatio']
data1norm['DebtToEquity Ratio'] = data1['DebtToEquity Ratio']
data1norm['TTMNetIncome'] = data1['TTMNetIncome']
data1norm['LTInvestmentsDet'] = data1['LTInvestmentsDet']
data1norm['TotalAssets'] = data1['TotalAssets']
data1norm['CashOnHand'] = data1['CashOnHand']
data1norm['LongTermDebt'] = data1['LongTermDebt']
data1norm['TotalLiabilities'] = data1['TotalLiabilities']
```
