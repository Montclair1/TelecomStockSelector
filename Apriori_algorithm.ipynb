```python
#importing some important libraries
from apyori import apriori
import pandas as pd
```

```python
#storing data to a variable (data frame) 
store_data = pd.read_csv('VerizonCat.csv', header=None)  
```

```python
#Display the first 5 lines of the data
store_data.head()
```

```python
#store all the information to a list so it gets easier to apply the algorithm 
records = []  
for i in range(0, 46):  
    records.append([str(store_data.values[i,j]) for j in range(0, 7)])
```

```python
#get some association rules 
association_rules = apriori(records, min_support=0.1, min_confidence=0.2,min_lift=1, min_length=2)
```

```python
#create a new list called association_results
association_results = [] 
```

```python
#convert the association_rules content to a list
association_results = list(association_rules)
```

```python
#print the length of the list, to see the number of the association_results
#Changing the values of min_support and min_confidence will have a huge affect on this value.
print(len(association_results))
```

```python
#display the results using a for loop
for item in association_results:
    print(item)
    print("")
```
